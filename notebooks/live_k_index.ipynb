# Live K Index
In this tutorial we demonstrate the live K index functionality.

- We make use of the live_k and plot_k functions to quickly load and plot the most up to date version of the k index for the chosen site. We use 'dun' for Dunsink in this example.

- We set the time using pandas to get the current time other times can be used to plot historic live K index plots as they would have been seen at the time

- live_k returns a dataframe with the live K index values and plot_k produces a simple bar chart from the dataframe returning the figure and subplot axes for further manipulation

```python
from magie import live_k, plot_k
import pandas as pd
now_time= pd.Timestamp.now()
plot_k(live_k(now_time, 'dun'))
```
