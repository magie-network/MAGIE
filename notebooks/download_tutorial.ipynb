# Tutorial for Download Code
Here we demonstrate how to use the download code

download_magie requires a start and end time it will then download all data from the provided list of sites as daily files and store them in a single hdf5 file under they key 'main'.
- Sites defaults to all sites
- save_file_name defaults to a local file containing the time range and site choices in the filename

```python
from magie import download_magie # import download function
import numpy as np
download_magie(np.datetime64('2022-01-01T00:00'), np.datetime64('2022-01-10T00:00'), 
                            sites=['dun'], save_file_name='../Data/Dunsink.hdf5')
```
