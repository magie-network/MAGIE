# Provisional K Tutorial
Here we demonstrate how to calculate the provisional K index for a site.
We use the same file downloaded in the download_tutorial.ipynb.

provisional_k can handle a range of inputs: dataframe, csv path, txt path and hdf5 path. For the case of hdf5 the code will chunk through the dataset allowing datasets of any size to be used without encountering memory issues. hr3_chunks decides the chunksize if a hdf5 path is provided in which case its the multiple of 3 hours that are loaded and processed at a time. The larger the value the quicker it runs however the more memory intensive it becomes.

If a hdf5 file is not provided as input then the output is a dataframe containing provisional K values

```python
from magie import provisional_k
provisional_k('../Data/Dunsink.hdf5',
            inkey='main',
                    outfile='../Data/Dunsink.hdf5',
                    outkey='provisional_K', hr3_chunks=5)
```

## Plotting Provisional K
We can simply use the plot_k function to plot the newly calculated provisional K index.
With a pandas hdf file we can search for the time range we are interested in

```python
from magie import plot_k
import numpy as np
import pandas as pd
start, end = np.datetime64('2022-01-02T00:00'), np.datetime64('2022-01-06T00:00')
df= pd.read_hdf('../Data/Dunsink.hdf5', key='provisional_K',
                where='Date_UTC>=start&Date_UTC<=end')
plot_k(df)
```

We can also plot the full time range if we wish
